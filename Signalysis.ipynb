```python
π = 3.1415926535897932384626433
```

```python
def min_sample(B):
    return B/π
```

```python
freq = [0.22,0.65]
list(map(min_sample, freq))
```

```python
modulus = lambda x, n : x%n
```

```python
x = range(2,16001)
y = [modulus(16000,i) for i in x]
```

```python
plt.scatter(x,y, s=.1)
plt.show()
```
